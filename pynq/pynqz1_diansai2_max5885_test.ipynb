# PYNQ-Z1 MAX5885 200 MSPS Bring-up

Dual MAX5885 overlay: board IO plus 200 MSPS waveform generator.


```python
from time import sleep, time
from pynq import Overlay, MMIO
BITFILE = 'base_add.bit'
overlay = Overlay(BITFILE)
LED_BASE, MAX5885_BASE = 0x40000000, 0x40001000
led_ip, max5885_ip = MMIO(LED_BASE, 0x1000), MMIO(MAX5885_BASE, 0x1000)
print('Overlay:', BITFILE)
print('IP:', sorted(overlay.ip_dict.keys()))
print('MAX5885 version=0x%08X, sample_rate=%d' % (max5885_ip.read(0x38), max5885_ip.read(0x3C)))
```

```python
# Reuse the verified physical LED/RGB/button test cell from the AD9767 notebook unchanged.
# It uses led_ip at 0x40000000 and remains compatible with this overlay.
```

```python
from pynqz1_diansai2_max5885 import MAX5885Signal
max5885 = MAX5885Signal(MAX5885_BASE)
cfg = max5885.configure_wireless(
    carrier_hz=30_000_000, mod_hz=2_000_000, square_hz=1_000_000,
    mode='AM', sd_vrms=0.25, am_depth_percent=30,
    sm_delay_ns=80, sm_phase_deg=30, sm_atten_db=6,
    out_a='SD', out_b='SQUARE')
print('MAX5885 config:', cfg)
print('MAX5885 status:', max5885.status())
```
