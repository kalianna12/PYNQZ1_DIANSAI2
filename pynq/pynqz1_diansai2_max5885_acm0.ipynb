# MAX5885 UART PL Control (replaces ACM0)

PL AXI UART 16550 on right expansion header W9(RX)/W10(TX), 115200 8E1.

Baud rate, parity, and all serial parameters are PS runtime-configurable via standard termios/pyserial.
The STM32 connects its UART1 TX to W9 and UART1 RX to W10.

Deploy these files into the same Jupyter directory before running:
- base_add.bit and base_add.hwh
- pynqz1_diansai2_max5885.py
- max5885_uart_listener_service.py

Each valid STM32 WCFG command immediately configures the real MAX5885 and pulses physical LD1 for 500 ms.


```python
# Cell 1: load the overlay and initialize the fixed ACM0 listener.
from pathlib import Path
from max5885_uart_listener_service import Max5885UartListener

APP_DIR = Path.cwd()
UART_PORT = '/dev/ttyACM0'

listener = Max5885UartListener(
    bitfile=APP_DIR / 'base_add.bit',
    uart_port=UART_PORT,
)
listener.initialize()
print('Ready: %s, 115200 8N1' % UART_PORT)
print('Valid command: WCFG,fc_hz,mode,sd_mv,sd_phase,am_depth,sm_delay,sm_phase,sm_atten,out_a,out_b,mod_hz,square_hz')
```

```python
# Cell 2: automatic listener. Interrupt this cell in Jupyter to stop it.
print('Listening on /dev/ttyACM0. Each valid WCFG applies MAX5885 settings and pulses LD1 for 500 ms.')
listener.run_forever()
```
